```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```



# Report on Prompt Variations for OrderBot

The objective of this exercise was to experiment with different system prompts to observe how GPT adapts its responses in various contexts. Instead of a fast-food restaurant, I tested three new scenarios: a bookstore assistant, a travel agent, and a music festival ticket bot.

### Findings

1. **Bookstore OrderBot**

   * This version worked well. The assistant followed the structure of greeting, collecting the customer’s book order, offering extras like bookmarks and tote bags, and summarizing the final purchase.
   * GPT sometimes improvised by recommending famous book titles even though the prompt instructed it to stick strictly to the menu. This was a mild form of hallucination but still contextually relevant.

2. **Travel Agent OrderBot**

   * This was the most successful variation. GPT clearly guided the customer step-by-step, confirmed details such as flight class and add-ons, and summarized prices accurately.
   * Rarely, GPT invented additional destinations (e.g., London) that were not in the menu. This showed that if the prompt was not explicit enough (“stick only to these four cities”), GPT tended to over-generalize.

3. **Music Festival OrderBot**

   * This context produced lively and engaging responses, especially when asked about tickets and merch. GPT often added emojis and an enthusiastic tone, which matched the festival theme well.
   * However, GPT sometimes treated the order like a conversation about music taste, asking irrelevant questions about favorite bands. While entertaining, it showed that the prompt could drift from the ordering task.

### Variations That Didn’t Work Well

The main problem across versions was **hallucination**: GPT occasionally added items not listed (book titles, cities, or food items). The more “open-ended” or creative the context, the more likely it was to drift away from the defined menu. In contrast, the tightly structured travel agent prompt kept GPT closer to the intended task.

### Lessons Learned

* **Prompt specificity matters.** The clearer and stricter the system instructions, the less GPT invents outside the menu.
* **Tone changes behavior.** A playful or thematic style (festival, romantic, detective) encouraged more creativity but also more drift. A formal/structured style kept the responses on-task.
* **Balance is key.** If the goal is accurate ordering, structured prompts are better. If the goal is customer engagement, playful prompts can make the interaction more fun but risk minor inaccuracies.

### Conclusion

By experimenting with different prompts and contexts, I learned that GPT is highly adaptable but also sensitive to prompt framing. Small changes in tone or detail can shift the conversation significantly, affecting both accuracy and style. For real-world applications like ordering systems, combining structure with some personality seems to deliver the best overall experience.
